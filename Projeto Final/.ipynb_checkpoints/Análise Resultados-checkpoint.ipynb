# Resultados Heurística + Local Search + Meta-Heurística

**1.** Variáveis macro:

    * Local Search (sim/nao): se há presença ou não de etapa extra de Local Search no processo de Iterated Greedy
    * Break (sim/nao): se o loop é interrompido ou não caso seja encontrado uma melhoria na etapa de Local Search


**2.** Variáveis micro:

    * T = 1
    * d = 4
    * iterações (IG) = 100

# 1. Leitura do arquivo com resultados

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(35,30)}, font_scale = 2)
```

```python
df = pd.read_csv('pandas_TT.csv')
```

```python
df.head()
```

# 2. Valores médios

```python
aux = df[df['instancia'] == 1].groupby(['LS', 'break']).mean()
aux
```

Realizando uma análise superficial a partir dos valores médios, verifica-se que a presença do break no Local Search influencia tanto o valor final de _Total Tardiness_ quanto o tempo de execução do algoritmo, de modo que, na presença do break caso encontre uma melhora, o _Total Tardiness_ final possui um valor pior e o tempo de execução do algoritmo é menor. 

Em relação a presença de mais uma etapa de _Local Search_ no _Iterated Greedy_ , verifica-se que há uma melhora significativa no valor do _Total Tardiness_ final, enquanto o tempo de execução cresce substancialmente.

# 5. Influência entre variáveis (Local Search e "break")

São criadas classes para avaliar tais influências:

| Classe |  LS | break |
|:------:|:---:|:-----:|
|    1   | sim |  sim  |
|    2   | sim |  nao  |
|    3   | nao |  sim  |
|    4   | nao |  nao  |

```python
LS_sim = df['LS'] == 'sim'
LS_nao = df['LS'] == 'nao'
break_sim = df['break'] == 'sim'
break_nao = df['break'] == 'nao'
```

```python
df.loc[(LS_sim) & (break_sim), 'classe'] = 1
df.loc[(LS_sim) & (break_nao), 'classe'] = 2
df.loc[(LS_nao) & (break_sim), 'classe'] = 3
df.loc[(LS_nao) & (break_nao), 'classe'] = 4

df['classe'].isna().sum()
```

## 5.1. Gráficos _Total Tardiness_

```python
inst1 = df.loc[df['instancia'] == 1, :]
inst2 = df.loc[df['instancia'] == 2, :]
inst3 = df.loc[df['instancia'] == 3, :]
inst4 = df.loc[df['instancia'] == 4, :]
inst5 = df.loc[df['instancia'] == 5, :]
inst6 = df.loc[df['instancia'] == 6, :]
inst7 = df.loc[df['instancia'] == 7, :]
inst8 = df.loc[df['instancia'] == 8, :]
inst9 = df.loc[df['instancia'] == 9, :]
inst10 = df.loc[df['instancia'] == 10, :]
```

```python
fig, axes = plt.subplots(5, 2)
_ = sns.boxplot(x = 'instancia', y = 'TT', data = inst1, hue = 'classe', ax = axes[0][0])

_ = sns.boxplot(x = 'instancia', y = 'TT', data = inst2, hue = 'classe', ax = axes[0][1])

_ = sns.boxplot(x = 'instancia', y = 'TT', data = inst3, hue = 'classe', ax = axes[1][0])

_ = sns.boxplot(x = 'instancia', y = 'TT', data = inst4, hue = 'classe', ax = axes[1][1])

_ = sns.boxplot(x = 'instancia', y = 'TT', data = inst5, hue = 'classe', ax = axes[2][0])

_ = sns.boxplot(x = 'instancia', y = 'TT', data = inst6, hue = 'classe', ax = axes[2][1])

_ = sns.boxplot(x = 'instancia', y = 'TT', data = inst7, hue = 'classe', ax = axes[3][0])

_ = sns.boxplot(x = 'instancia', y = 'TT', data = inst8, hue = 'classe', ax = axes[3][1])

_ = sns.boxplot(x = 'instancia', y = 'TT', data = inst9, hue = 'classe', ax = axes[4][0])

_ = sns.boxplot(x = 'instancia', y = 'TT', data = inst10, hue = 'classe', ax = axes[4][1])
```

Verifica-se que, na presença de outra etapa de Local Search, a presença de "break" não possui grande influência, sendo os resultados bastante semelhantes. 

Já na ausência da etapa de Local Search, o "break" demonstra alta variabilidade de valores de _Total Tardiness_ , como observado nas instâncias 4, 5, 6 e 10.

## 5.2. Gráficos Tempo

```python
fig, axes = plt.subplots(5, 2)
_ = sns.boxplot(x = 'instancia', y = 'tempo', data = inst1, hue = 'classe', ax = axes[0][0])

_ = sns.boxplot(x = 'instancia', y = 'tempo', data = inst2, hue = 'classe', ax = axes[0][1])

_ = sns.boxplot(x = 'instancia', y = 'tempo', data = inst3, hue = 'classe', ax = axes[1][0])

_ = sns.boxplot(x = 'instancia', y = 'tempo', data = inst4, hue = 'classe', ax = axes[1][1])

_ = sns.boxplot(x = 'instancia', y = 'tempo', data = inst5, hue = 'classe', ax = axes[2][0])

_ = sns.boxplot(x = 'instancia', y = 'tempo', data = inst6, hue = 'classe', ax = axes[2][1])

_ = sns.boxplot(x = 'instancia', y = 'tempo', data = inst7, hue = 'classe', ax = axes[3][0])

_ = sns.boxplot(x = 'instancia', y = 'tempo', data = inst8, hue = 'classe', ax = axes[3][1])

_ = sns.boxplot(x = 'instancia', y = 'tempo', data = inst9, hue = 'classe', ax = axes[4][0])

_ = sns.boxplot(x = 'instancia', y = 'tempo', data = inst10, hue = 'classe', ax = axes[4][1])
```
