Importando as bibliotecas necessárias

```python
import xlwings as xw
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
```

Definindo a classe Job que usaremos no algoritmo

```python
class Job:
    def __init__(self,i,pi,di,Ci):
        self.i=i     # número do job, pela ordem de chegada
        self.p=pi     # processing time
        self.p_total=sum(self.p)
        self.d=di     # due date
        self.C=Ci     # completion time
        self.T=0     # tardiness
        self.bestPos=-1 # Esse novo atribuito tem o objetivo de guardar a melhor posição do job, para a ordenação posterior
    
    def __str__(self):
        return 'Job #{}, p = {}'.format(self.i, self.p)
```

### Reutilizamos as funções disponibilizada pelo professor, com alguns ajustes para ler o número de máquinas e gravar o Runtime e Gap

```python
def leInst(sheet_num):
    pl1 = wb1.sheets[sheet_num]
    m=int(pl1.range('B5').value)     # converter em inteiro
    n=int(pl1.range('B6').value)     # converter em inteiro
    Cmax=int(pl1.range('F5').value)     # converter em inteiro
    tab1=pl1.range('A11').options(expand='table', numbers=int).value     # ler como tabela e inteiros
    jobs=[]
    for i in range(n):
        pi=tab1[i][1:-5]
        Ci=tab1[i][-5:]
        jobs.append(Job(tab1[i][0]-1,pi,0,Ci))
    
    return m,n,jobs,Cmax
```

.

### Função que faz o Plot do gráfico de Gantt para qualquer lista de jobs

```python
def graphGantt(jobs: list, m: int = 5):
    
    # Cores para as barras de cada job
    colors = np.array(plt.get_cmap('tab20').colors)
    np.random.shuffle(colors)
    
    fig, ax = plt.subplots()

    # Colocando títulos nos eixos
    ax.set_xlabel('Tempo') 
    ax.set_ylabel('Máquina')
    
    y = 5   # Variável que representa a altura da barra no gráfico
    height = 9   # Comprimento das barras no eixo y
    ticks = []   # Lista que guarda a coordenada y das barras plotadas, para colocar legenda depois
    labels = []   # Lista que guarda os nomes das Máquinas, para a legenda

    for maq in range(m-1, -1, -1):
        
        lista_start_p = []
        for job in jobs:
            lista_start_p.append((job.C[maq]-job.p[maq], job.p[maq]))   # Cria lista de tuplas: [(start_time, p_time) ...]

        # Plotando uma única máquina, com todos os jobs para ela
        ax.broken_barh(lista_start_p, (y, height), color=colors)
        
        ticks.append(y+height/2)
        labels.append(f'M{maq+1}')
        
        y += 10
        
    plt.yticks(ticks=ticks, labels=labels)   # Coloca as legendas nos seus lugares

    plt.show()   # Imprime o plot
```
