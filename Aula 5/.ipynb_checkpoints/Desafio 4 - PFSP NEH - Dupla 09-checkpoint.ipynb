Definindo a classe Job que usaremos no algoritmo

```python
class Job:
    def __init__(self,i,pi,di):
        self.i=i     # número do job, pela ordem de chegada
        self.p=pi     # processing time
        self.p_total=sum(self.p)
        self.d=di     # due date
        self.C=0     # completion time
        self.T=0     # tardiness
```

Função que permite calcular o tempo que os jobs se completam

```python
def calcCmax(m,jobs):     # esta função calcula os tempos de conclusão dos jobs nas m máquinas
    n=len(jobs)
    c=[[0 for i in range(m)] for j in range(n)]
    p=[job.p for job in jobs]
    
    # primeiro job
    c[0][0]=p[0][0]     # primeira máquina
    for k in range(1,m):     # próximas máquinas
        c[0][k]=c[0][k-1]+p[0][k]
        
    # próximos jobs
    for j in range(1,n):
        c[j][0]=c[j-1][0]+p[j][0]     # primeira máquina
        for k in range(1,m):     # próximas máquinas
            c[j][k]=max(c[j][k-1],c[j-1][k])+p[j][k]
            
    j=0
    for job in jobs:
        job.C=c[j]
        j+=1
    
    return c[-1][-1]
```

### Procedimento
    
        1. Para cada job `i` calcular: 
        
                1.2. `Ti = SUM(t(i,j))` para j=1,...,m
    
        2. Ordenar os jobs em ordem decrescente de `Ti`
    
        3. Escolher os dois primeiros jobs da lista ordenada e ache a melhor sequência entre as possibilidades. Não mude suas posições relativas nos próximos passos do algoritmo.

        4. Para i=3 até n:

                4.1. Escolha o job na posição `i` da lista gerada no Passo 2 e ache a melhor sequência colocando este job em todas as posições possíveis `i` da sequência parcial do passo anterior.

### Função `permutar` é responsável por achar a posição qual o job sendo inserido na lista resulta no menor _makespan_  possível

```python
def permutar(jobs, new_job, m):

    bestCmax = 9999999999999

    for index in range(len(jobs)+1):

        aux_jobs = jobs[:]

        aux_jobs.insert(index, new_job)

        Cmax = calcCmax(m, aux_jobs)

        if Cmax < bestCmax:
            final_jobs = aux_jobs
            bestCmax = Cmax
    
    bestCmax = calcCmax(m, final_jobs)

    return final_jobs, bestCmax
```

### Função principal `schedNEH` ordena a lista na ordem certa e insere os itens 1 a 1

```python
def schedNEH(jobs, m):

    jobs.sort(key = lambda job: job.p_total, reverse=True)

    jobs_permutacao = jobs[:1]

    for i in range(1, len(jobs)):

        jobs_permutacao, Cmax = permutar(jobs_permutacao, jobs[i], m)
    
    return jobs_permutacao,Cmax
```
