## 2. Análise de Resultados

**Tabela de Variância**

```r
av <- aov(Cmax~d*T)
summary(av)
```

**Boxplots**

* Relação entre Cmax e d:

```r
boxplot(Cmax ~ d)
```

* Relação entre Cmax e Temperatura:

```r
boxplot(Cmax ~ T)
```

* Relações combinadas entre Cmax e d e Temperatura

```r
boxplot(Cmax ~ d*T)
```

# 3. Conclusão

Verifica-se, então, que tanto a variável d quanto a temperatura não influenciam significativamente nos resultados finais de _makespan_, observando que as medianas de todos _boxplots_ pouco se alteram, independente da variável em teste. Para o gráfico de influência combinada, os inter-quartis se alteram, mas a mediana continua estável. Isto pode ser explicado pelo fato de que o método _Iterated Greedy_ (IG) é o último aplicado, após o _NEH_ e o _Local Search_ (LS), que já fornecem bons resultados. Assim, não há grandes melhorias a serem executadas nos dados, culminando, pois, em uma pequena influência das variáveis d e temperatura em relação aos resultados.
